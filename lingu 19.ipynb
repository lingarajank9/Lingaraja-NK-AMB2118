```python
s = input("Enter a string: ")
rev = ""

for ch in s:
    rev = ch + rev

print("Reversed string:", rev)
```

```python
s = input("Enter a string: ")
rev = s[::-1]

if s == rev:
    print("Palindrome")
else:
    print("Not a Palindrome")
```
