```python
#Assignment Operators

a = 10
print("Initial value:", a)

a += 5				#a=a+5
print("After += :", a)		#a=10+5=15

a -= 3				#a=a-3
print("After -= :", a)		#a=15-3=12

a *= 2				#a=a*2
print("After *= :", a)		#a=12*2=24

a /= 4				#a=a/4
print("After /= :", a)		#a=24/4=6

a //= 2				#a=a/2 (floor value)
print("After //= :", a)		#a=6/2=3

a %= 3				#a=a/3 (remainder)
print("After %= :", a)		#a=3/3 (remainder=0)

a **= 2				#a=a2
print("After **= :", a)		#a=02 = 0
```
