```python
import tkinter as tk
from PIL import Image, ImageTk

# Create window
window = tk.Tk()
window.title("Food & Drinks")
window.geometry("900x700")

# Food list (Name, Image file)
foods = [
    ("Burger", "burger.jpg"),
    ("Pizza", "pizza.jpg"),
    ("Butter", "butter.jpg"),
    ("Curd", "curd.jpg"),
    ("Milk", "milk.jpg"),
    ("Dosa", "dosa.jpg"),
    ("Tea", "tea.jpg"),
    ("Coffee", "coffee.jpg"),
    ("Lemon Juice", "lemon_juice.jpg"),
    ("Orange Juice", "orange_juice.jpg"),
    ("Pomegranate Juice", "pomegranate_juice.jpg")
]

images = []  # keep reference to images

row = 0
col = 0

for name, img_file in foods:
    try:
        img = Image.open(img_file)
        img = img.resize((120, 120))
        photo = ImageTk.PhotoImage(img)
        images.append(photo)

        label_img = tk.Label(window, image=photo)
        label_img.grid(row=row, column=col, padx=10, pady=10)

        label_text = tk.Label(window, text=name, font=("Arial", 12))
        label_text.grid(row=row+1, column=col)

        col += 1
        if col == 4:
            col = 0
            row += 2

    except:
        print(f"Image not found: {img_file}")

window.mainloop()
```

```python
cgpa = float(input("Enter CGPA: "))
backlogs = int(input("Enter number of backlogs: "))
semester = int(input("Enter current semester: "))

if cgpa > 8 and backlogs == 0 and semester == 8:
    print("Student eligible for placement")
else:
    print("Student not eligible for placement")
```
